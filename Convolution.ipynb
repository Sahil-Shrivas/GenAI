```python
import tensorflow as tf
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
import requests
import io

# Define the image URL
image_url = "https://picsum.photos/256/256" # Changed to a publicly accessible image URL

# Download the image content
response = requests.get(image_url)
response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

# Load an image directly from bytes using PIL
image = Image.open(io.BytesIO(response.content)).convert("L")  # Load as grayscale
image = image.resize((256, 256))  # Resize to make easy
image = np.array(image, dtype=np.float32) / 255.0  # Convert to float32 and normalize
# Add batch and channel dimensions (1, 256, 256, 1)
image = np.expand_dims(image, axis=(0, -1))

# Define vertical and horizontal edge detection filters (Sobel-like)
vertical_filter = np.array([[1, 0, -1],
                            [1, 0, -1],
                            [1, 0, -1]], dtype=np.float32)

h_filter = np.array([[1,  1,  1],
                     [0,  0,  0],
                     [-1, -1, -1]], dtype=np.float32)

# Reshape filters for TensorFlow (H, W, in_channels, out_channels)
vertical_filter = vertical_filter.reshape(3, 3, 1, 1)
h_filter = h_filter.reshape(3, 3, 1, 1)

# Apply convolution
vertical_edges = tf.nn.conv2d(image, vertical_filter, strides=[1, 1, 1, 1], padding="SAME")
h_edges = tf.nn.conv2d(image, h_filter, strides=[1, 1, 1, 1], padding="SAME")

# Convert tensors to numpy for visualization
vertical_edges = vertical_edges.numpy().squeeze()
horizontal_edges = h_edges.numpy().squeeze()



```

```python
# Plot original and filtered images
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(image.squeeze(), cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(vertical_edges, cmap="gray")
plt.title("Vertical Edges")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(horizontal_edges, cmap="gray")
plt.title("Horizontal Edges")
plt.axis("off")

plt.show()
```
